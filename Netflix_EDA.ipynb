```python
1. Bar chart → Count of Movies vs TV Shows
```

```python
type_counts = df['type'].value_counts()

plt.figure(figsize=(10,6))
plt.bar(type_counts.index,type_counts.values,width=0.2,color=['green','red'])
plt.xlabel(' Show Type')
plt.ylabel('Count')
plt.title('Tv Shows Vs Movies')
plt.tight_layout()
plt.show()
```

```python
2. Pie chart → Top 5 countries by number of titles
```

```python
titles_counts=df['country'].value_counts().head(5)

plt.figure(figsize=(10,6))
plt.pie(titles_counts.values,labels=titles_counts.index,radius=1,autopct='%1.1f%%')
plt.title('Top 5 countries by number of titles')
```

```python
3.  Bar chart → Top 5 genres
```

```python
genre_counts = df['listed_in'].value_counts().head(5)

plt.figure(figsize=(10,6))
plt.bar(genre_counts.index,genre_counts.values,width=0.4)
plt.xlabel('Genres')
plt.ylabel('Count')
plt.title('Top 5 genres')
plt.xticks(rotation=10)
plt.tight_layout()
```

```python
4. Line chart → Number of releases by year
```

```python
year_counts = df['release_year'].value_counts().sort_index()

plt.figure(figsize=(10,6))
plt.plot(year_counts.index,year_counts.values,color='green')
plt.xlabel('Years')
o=plt.ylabel('No.of Releases')
plt.title('Number of releases by year')
```

```python
5. Histogram → Distribution of release years
```

```python
plt.figure(figsize=(10,5))
plt.hist(df['release_year'],bins=20,color='purple',edgecolor='black')
plt.xlabel('Years')
plt.ylabel('No.of Releases')
plt.title('Release Year Distribution')
plt.tight_layout()
```
